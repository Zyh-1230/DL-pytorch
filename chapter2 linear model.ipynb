```python
import numpy as np
import matplotlib.pyplot as plt
plt.grid = plt.matplotlib.pyplot.grid
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
```

```python
def forward(x):
    return x * w
```

```python
def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) ** 2
```

```python
w_list = []
mse_list = []
```

```python
for w in np.arange(0.0, 4.1, 0.1):
    print('w=', w)
    loss_sum = 0
    for x_val, y_val in zip(x_data, y_data):
        y_pred_val = forward(x_val)
        loss_val = loss(x_val, y_val)
        loss_sum += loss_val
        print('\t', x_val, y_val, y_pred_val, loss_val)
    print('MSE=', loss_sum/3)
    w_list.append(w)
    mse_list.append(loss_sum / 3)
```

```python
plt.figure()
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.grid(True)
plt.show()
```
