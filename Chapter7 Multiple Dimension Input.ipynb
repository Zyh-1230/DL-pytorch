```python
import torch
import numpy as np
x = np.loadtxt('dataset/diabetes_data/diabetes_data_raw.csv.gz', delimiter=' ', dtype=np.float32)
y = np.loadtxt('dataset/diabetes_data/diabetes_target.csv.gz', dtype=np.float32)
x_data = torch.from_numpy(x)
y_data = torch.from_numpy(y).view(-1, 1) # 将一维张量转换为二维列向量
```

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear1 = torch.nn.Linear(10, 8) # !!!
        self.linear2 = torch.nn.Linear(8, 6)
        self.linear3 = torch.nn.Linear(6, 4)
        self.linear4 = torch.nn.Linear(4, 1)
        self.sigmoid = torch.nn.Sigmoid()
        self.activate = torch.nn.ReLU()

    def forward(self, x):
        x = self.sigmoid(self.linear1(x))
        x = self.activate(self.linear2(x))
        x = self.sigmoid(self.linear3(x))
        x = self.linear4(x) # 回归问题需要线性输出
        return x

model = Model()
```

```python
criterion = torch.nn.MSELoss(reduction='mean') # 数据集为实数 非0-1二分类 故采用MSELoss 而非BCELoss

optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(10000):
    y_pred= model(x_data)
    loss = criterion(y_pred, y_data)
    print(epoch, loss.item())

    optimizer.zero_grad()
    loss.backward()

    optimizer.step()
```
