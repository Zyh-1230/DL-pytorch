```python
import numpy as np
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
BATCH_SIZE = 64
NUM_EPOCH = 200
```

```python
torch.manual_seed(42)
np.random.seed(42)
```

```python
class TitanicDataset(Dataset):
    def __init__(self, is_train_set=True):
        self.is_train_set = is_train_set
        filename = 'train.csv' if is_train_set else 'test.csv'
        self.df = pd.read_csv(filename)

        if not is_train_set:
            self.passenger_ids = self.df['PassengerId'].copy()

        self._preprocess()

        self.features = torch.tensor(self.df.drop('Survived', axis=1).values, dtype=torch.float32) # 将根据列名取出的value转换为torch.float32标准单精度浮点数
        self.labels = torch.tensor(self.df['Survived'].values, dtype=torch.long) # torch.long = torch.int64
        self.len = len(self.df) # DataLoader以此确定数据总样本量、批量迭代次数、进度显示

    def _preprocess(self):
        self.df = self.df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

        self.df['Age'] = self.df['Age'].fillna(self.df['Age'].median())
        self.df['Embarked'] = self.df['Embarked'].fillna(self.df['Embarked'].mode()[0]) # mode()返回数据集众数的序列

        self.df['Sex'] = self.df['Sex'].map({'female': 1, 'male': 0})
        self.df['Embarked'] = self.df['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})

        self.df['Family_num'] = self.df['SibSp'] + self.df['Parch']
        self.df = self.df.drop(['SibSp', 'Parch'], axis=1)

        self.df = (self.df - self.df.mean()) / self.df.std() # 数据归一化

        if 'Survived' not in self.df.columns:
            self.df['Survived'] = -1 # 如果是测试集且没有标签列 添加一个占位符 占位符值为-1

    def __getitem__(self, index):
        return self.features[index], self.labels[index]

    def __len__(self):
        return self.len
```

```python
trainset = TitanicDataset(is_train_set=True)
train_loader = DataLoader(trainset, batch_size=BATCH_SIZE, shuffle=True)

testset = TitanicDataset(is_train_set=False)
test_loader = DataLoader(testset, batch_size=BATCH_SIZE, shuffle=False)
```

```python
class Model(torch.nn.Module):
    def __init__(self, input_size):
        super(Model, self).__init__()
        self.fc1 = torch.nn.Linear(input_size, 128) # 低维->高维->低维：增加维度创建交互的特征空间 
        self.fc2 = torch.nn.Linear(128, 64)
        self.fc3 = torch.nn.Linear(64, 32)
        self.fc4 = torch.nn.Linear(32, 1)
        self.relu = torch.nn.ReLU()
        self.dropout = torch.nn.Dropout(0.3) # Dropout层 丢弃概率为30%  防止过拟合(随机关闭部分神经元) 增强泛化能力
        self.sigmoid = torch.nn.Sigmoid()
        
        self.sigmoid = torch.nn.Sigmoid()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.relu(self.fc3(x))
        x = self.dropout(x)
        x = self.sigmoid(self.fc4(x))
        
        return x

input_size = trainset.features.shape[1] # trainset.features.shape返回torch.Size([891, 7])
model = Model(input_size)
```

```python
criterion = torch.nn.BCELoss(reduction='mean')
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
# 'min': 监控损失最小化 patience: 等待5个epoch无改善 factor: 学习率减半
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=5, factor=0.5)

train_losses = []
val_accuracies = []
```

```python
for epoch in range(NUM_EPOCH):
    model.train()
    running_loss = 0.0

    for i, (inputs, labels) in enumerate(train_loader): # train_loader返回批量的数据和对应的标签(inputs, labels) inputs:batch_size x num_features labels:batch_size x 1
        outputs = model(inputs).squeeze() # 输出的原始形状为[batch_size, 1] 经过squeeze后(去除冗余维度 与标签labels的维度完全匹配) [batch_size]
        loss = criterion(outputs, labels.float()) # 要求outputs.shape == labels.shape

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    epoch_loss = running_loss / len(train_loader)
    train_losses.append(epoch_loss)

    model.eval() # 在训练集上评估模型
    correct = 0
    total = 0

    with torch.no_grad():
        for inputs, labels in train_loader:
            outputs = model(inputs).squeeze()
            predicted = (outputs >= 0.5).float()
            total += labels.size(0)
            correct += (predicted == labels.float()).sum().item()

    train_accuracy = 100 * correct / total

    scheduler.step(epoch_loss)

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{NUM_EPOCH}], Loss: {epoch_loss:.4f}, Train Acc: {train_accuracy: .2f}%')

print('Finished Training')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(train_losses, label='Training Loss')
plt.title('Training Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.show()
```

```python
model.eval()
all_predictions = []
test_ids = []

with torch.no_grad():
    for inputs, _ in test_loader:
        outputs = model(inputs).squeeze()
        predicted = (outputs >= 0.5).int()
        all_predictions.extend(predicted.tolist())

    test_ids = testset.passenger_ids.tolist()
```

```python
submission = pd.DataFrame({'PassengerId': test_ids, 'Survived': all_predictions})

submission.to_csv('titanic_submission.csv', index=False)
print("预测结果已存为 'titanic_submission.csv'")

print("\n测试集前10个预测结果:")
print(submission.head(10))
```
