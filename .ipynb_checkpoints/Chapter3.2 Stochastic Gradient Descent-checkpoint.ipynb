```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
```

```python
w = 1.0
```

```python
def forward(x):
    return x * w

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) ** 2 # 求单个点的loss 而非所有点的loss

def gradient(x, y):
    return 2 * x * (w * x - y) # 求单个点的梯度 非所有点的梯度
```

```python
print('Predict (before training)', 4, forward(4))

for epoch in range(100):
    for x, y in zip(x_data, y_data):
        grad = gradient(x, y)
        w = w - 0.01 * grad # 不同于梯度下降 计算下一个w时依赖于上一个w 因此相邻数据点的计算不能并行 运行时间会加长
        print("\tgrad: ", x, y, grad)
        l = loss(x, y)
    print("Progress:", epoch, "w=", w, "loss=", l)

print('Predict (after training)', 4, forward(4))
```
